### PREPROCESING

```python
from sklearn.preprocessing import StandardScaler
X = pima.iloc[:,0:8]
Y = pima.iloc[:,8]
rescaledX = StandardScaler().fit_transform(X)
X_nuevo = pd.DataFrame(data = rescaledX, columns= X.columns)
X_nuevo.head()
```

```python
Y.head()
```

### NAIVE BAYES

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, random_state = 22, test_size = 0.2)
```

```python
models = []
models.append(("Naive_Bayes_Gaussian",GaussianNB()))
```

```python
results = []
names = []
for name,model in models:
    kfold = KFold(n_splits=10, random_state=22)
    cv_result = cross_val_score(model,X_train,Y_train, cv = kfold,scoring = "accuracy")
    names.append(name)
    results.append(cv_result)
for i in range(len(names)):
    print(names[i],results[i].mean())
```

```python
ax = sns.boxplot(data=results)
ax.set_xticklabels(names)
plt.show()
```

```python
results
```

### Evaluating

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
naive = GaussianNB()
naive.fit(X_train,Y_train)
naive_predictions = naive.predict(X_test)

print(accuracy_score(Y_test,naive_predictions))
print(classification_report(Y_test,naive_predictions))
```

```python
conf = confusion_matrix(Y_test,naive_predictions)
label = ["0","1"]
sns.heatmap(conf, annot=True, xticklabels=label, yticklabels=label)
plt.show()
```

```python
Y.head()
```

## STANDARIZATION

```python
from sklearn.preprocessing import StandardScaler
rescaledX = StandardScaler().fit_transform(X_features)
```

```python
X = pd.DataFrame(data = rescaledX, columns= X_features.columns)
```

```python
X.head()
```

## BINARY CLASSIFICATION

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, random_state = 22, test_size = 0.2)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
models = []
models.append(("LR",LogisticRegression()))
models.append(("NB",GaussianNB()))
models.append(("KNN",KNeighborsClassifier()))
models.append(("DT",DecisionTreeClassifier()))
models.append(("SVM",SVC()))
```

```python
results = []
names = []
for name,model in models:
    kfold = KFold(n_splits=10, random_state=22)
    cv_result = cross_val_score(model,X_train,Y_train, cv = kfold,scoring = "accuracy")
    names.append(name)
    results.append(cv_result)
for i in range(len(names)):
    print(names[i],results[i].mean())
```

## VISUALISING RESULTS

```python
ax = sns.boxplot(data=results)
ax.set_xticklabels(names)
```
